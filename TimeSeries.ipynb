```python
import numpy as np
```

```python
anom = np.load('anomaly_samples.npy',allow_pickle=True).astype(np.float64)
norm = np.load('normal_samples.npy',allow_pickle = True).astype(np.float64)
anom[0:200].shape
```

```python
anom[0:200].shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
n_anom = anom.shape[0]
n_norm = norm.shape[0]

start = 1000
length = 2000


data = anom[start:start+length]  # Generating random data for demonstration

# Generating time indices (assuming each row represents a time point)
time = np.arange(length)

# Plotting each column as a time series in separate subplots (vertical layout)
fig, axs = plt.subplots(7, 1, figsize=(8, 12))  # 7 rows, 1 column

for i, ax in enumerate(axs):
    ax.plot(time, data[:, i], label=f'Feature {i+1}')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    ax.set_title(f'Feature {i+1}')
    ax.grid(True)
    ax.legend()

plt.tight_layout()  # Adjust subplots to prevent overlap
plt.show()
```
