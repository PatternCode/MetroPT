```python
# Load the MNIST dataset
from sklearn.datasets import fetch_openml
mnist = fetch_openml("mnist_784")

# Normalize pixel values to [0, 1]
X = mnist.data / 255.0    # Pandas data frame
y = mnist.target          # Pandas Series
```

```python
from sklearn.manifold import TSNE

# Initialize t-SNE with 2 components (2D visualization)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
X_tsne.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame for plotting
df_tsne = pd.DataFrame(X_tsne, columns=["tsne1", "tsne2"]) # Create a DataFram from a numpy array()
df_tsne["label"] = y.astype(int)   # Add a new column

# Plot the scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x="tsne1", y="tsne2", hue="label", palette="tab10", data=df_tsne)
plt.title("t-SNE Visualization of MNIST Dataset")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(title="Digit Class")
plt.show()
```
